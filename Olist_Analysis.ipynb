## 2.1 Recency : The customer transaction, by measuring that we can investigate the customer engagment with Olist business.

```python
import datetime as dt
time_limit = merged_df['order_purchase_timestamp'].max() + dt.timedelta(1)
rfm = merged_df.groupby('customer_unique_id', as_index=False).agg(
    recency=('order_purchase_timestamp', lambda x: (time_limit - x.max()).days),
    frequency=('order_id', 'count'),
    monetary=('payment_value', 'sum'),
)
```

## 2.4 K-Means Clustering

We want to rank the customers based on a certine scale, according to the distruibtion showed above we have outliers and that is not efficient for a standared scaler, K-means is a ml algorithim helps with that.
-  K-means clustering will be applied to the RFM scores of customer to segment
 them into distinct groups based on their behaviour and engagement.

```python
from sklearn.preprocessing import StandardScaler
# Measures
x = rfm[['recency', 'frequency', 'monetary']]
# Scaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Calculate Within Cluster Sum Squares (WCSS)
from sklearn.cluster import KMeans
WCSS = []
for i in range(2,9):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=11)
    kmeans.fit(x_scaled)
    WCSS.append(kmeans.inertia_)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

silhouette_scores = []
cluster_range = range(2, 8)  

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, max_iter=100, random_state=11, n_init='auto').fit(x_scaled)
    score = silhouette_score(x_scaled, kmeans.labels_)
    silhouette_scores.append(score)
    print(f"k={k}: Silhouette Score = {score:.4f}")

plt.figure()
plt.plot(cluster_range, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.show()

optimal_k = cluster_range[np.argmax(silhouette_scores)]
print(f"\nOptimal number of clusters (k): {optimal_k}")
```

```python
# Elbow Method
plt.figure()
plt.plot(range(2,9), WCSS)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

Find the point where adding more clusters doesnt significantly reduce WCSS (within-cluster variance). the elbow appears around 6 clusters (between 4 and 6), where the WCSS curve starts flattening.

The "elbow" (optimal point) is where the curve bendshere, it occurs around 4 cluster

```python
# 4 clusters are chosen 
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=11)
kmeans.fit(x_scaled)
rfm['label_kmeans'] = kmeans.labels_
```

*By analyzing these clusters, marketers can identify patterns and tailor their market campaign to meet the specific needs of each customer segment.*

| **Tier**          | **Behavior Summary**                          |
|-------------------|-----------------------------------------------|
| **New Customers** | Customers with low engagement who have transacted recently        |
| **Potential Loyal** | Customers showing loyalty signals (Recent customers but with a good amount of spending/frequency) | 
| **Loyal Customers** | Frequent buyers with high lifetime value (had the highest total transactions/monetary value)| 
| **Churn Risks**   | Inactive customers who may not return   (Have made a transaction while ago with low frequency/monetary purchases)|

```python
import numpy as np
price_tiers = [
    (0, 20, "Under R$20"),
    (20, 50, "R$20-R$50"), 
    (50, 150, "R$50-R$150"),
    (150, 500, "R$150-R$500"),
    (500, 2000, "R$500-R$2000"),
    (2000, np.inf, "R$2000+")
]

labels = [tier[2] for tier in price_tiers]
bins = [tier[0] for tier in price_tiers] + [price_tiers[-1][1]]
# Categorize prices
merged_df['price_tier'] = pd.cut(
    merged_df['price'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
price_bins = [0, 20, 50, 150, 500, np.inf]
price_labels = [
    'Under R$20', 
    'R$20-R$50',
    'R$50-R$150', 
    'R$150-R$500',
    'R$500-R$2000'
]

merged_df['price_tier'] = pd.cut(
    merged_df['price'],
    bins=price_bins,
    labels=price_labels,
    include_lowest=True
)
segment_price_preference = (merged_df
                          .groupby(['Customer Segment', 'price_tier'])
                          .size()
                          .unstack()
                          .apply(lambda x: x/x.sum(), axis=1))
plt.figure(figsize=(12, 8))
sns.heatmap(
    segment_price_preference,
    annot=True,
    fmt='.1%',
    cmap='YlGnBu',
    vmin=0,
    vmax=0.5,
    linewidths=0.5
)
plt.title('Customer Segment Price Preferences (BRL R$)', pad=20)
plt.xlabel('Price Tier')
plt.ylabel('Customer Segment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

preferred_tiers = segment_price_preference.idxmax(axis=1)
avg_spend = merged_df.groupby('Customer Segment')['price'].mean()

print("\nKey Insights:")
for segment in segment_price_preference.index:
    print(f"- {segment}: Prefers {preferred_tiers[segment]} tier (Avg spend: R${avg_spend[segment]:.2f})")
```
